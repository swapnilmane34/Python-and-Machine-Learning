convert categorial column value of diagonis into boolean ovalue
M = malignant or B = benign. 1 means the cancer is malignant and 0 means benign

```python
df['diagnosis']=pd.get_dummies(df['diagnosis'])
```

```python
# creating dataset
X = df.iloc[:,1:31]
Y = df.iloc[:, 31]
```

```python
X.head()
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
model=LogisticRegression()
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model.fit(X_train,Y_train)
```

```python
print('intercept b0',model.intercept_)
print('slope of line b1',model.coef_)
```

```python
y_pred=model.predict(X_test)
```
