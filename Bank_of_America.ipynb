```python
from pandas_datareader import data, wb
import pandas_datareader.data as web

import pandas as pd
import numpy as np
import datetime
```

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2016,1,1)
```

```python
BAC=data.DataReader('BAC','yahoo',start,end)
```

```python
BAC.head()
```

```python
BAC.columns
```

```python
BAC['High'].max()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.distplot(BAC['High'],color='r',kde=True,rug=True,label='High',bins=100)
plt.legend()
plt.subplot(2,2,2)
sns.distplot(BAC['Low'],color='m',kde=True,rug=True,label='Low',bins=100)
plt.legend()
plt.subplot(2,2,3)
sns.distplot(BAC['Open'],color='c',kde=True,rug=True,label='Open',bins=100)
plt.legend()
plt.subplot(2,2,4)
sns.distplot(BAC['Close'],color='g',kde=True,rug=True,label='Close',bins=100)
plt.legend()
```

```python
sns.scatterplot(x='High',y='Low',data=BAC,color='r')
```

```python
sns.distplot(BAC.loc['2008-01-01':'2008-12-31']['High'],color='m',bins=100)
```

```python
plt.figure(figsize=(10,6))
BAC['High'].loc['2006-1-1':'2008-1-1'].plot(color='r',label='High')
BAC['Close'].loc['2006-1-1':'2008-1-1'].rolling(window=30).mean().plot(label='mean')
plt.legend()
```

```python
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC CLOSE')
```
